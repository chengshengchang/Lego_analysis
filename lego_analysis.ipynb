```python
import pandas as pd
```

```python
# read csv 
lego_sets_df = pd.read_csv("./datasets/lego_sets.csv")
parent_theme_df = pd.read_csv("./datasets/parent_themes.csv")
```

```python
# merge dataframe and drop name_y cuz is as same as parent theme
merge_df = lego_sets_df.merge(parent_theme_df , left_on='parent_theme' , right_on='name')
merge_df.drop(columns='name_y' , inplace=True)
```

```python
# Task 1 : What is the most popular licensed sets ever were Star Wars Themed (Fiexd)
is_licensed_df = merge_df[merge_df['is_licensed']==True]
is_licensed_df = is_licensed_df.dropna(subset=['set_num'])
is_licensed_Starwar_df = is_licensed_df[is_licensed_df['parent_theme'] == 'Star Wars']

the_force =int(is_licensed_Starwar_df.shape[0] / is_licensed_df.shape[0]  * 100)
```

```python
# Task 2 : In which year was Star Wars not the most popular licensed theme ? 

licensed_sorted = is_licensed_df.sort_values('year')
licensed_sorted['count'] = 1

df_groupby = licensed_sorted.groupby(['year','parent_theme']).sum().reset_index()
df_groupby = df_groupby.sort_values('year')
df_groupby = df_groupby.sort_values('is_licensed',ascending=False).drop_duplicates(['year'])
df_groupby.sort_values('year',inplace=True)
df_groupby
```
